$$\textbf{Teleportation}$$

Teleportation is achieved by sharing two entangled qubits and with the help of a third one. Teleportation is about transporting the state of the third qubit from one entangled particle to the other entangled particle. Just information is transmitted, not the particle by itself.

Consider that Alices has qubit $A$ and Bob qubit $B$ that are entangled in the state $$| \phi^+ \rangle = \frac{1}{\sqrt{2}}|00 \rangle + \frac{1}{\sqrt{2}}|11 \rangle$$

Alice then earns a thir qubit $Q$ that wishes to transmit to Bob, i.e. she wishes Bob to have a qubit with the same information as $Q$, having whatever correlations that $Q$ had with other systems.

The quantum circuit that allows them to accomplish this taks is as follows:



The diagonal wires represent classical bits since measurements have already been performed. This classical bits are sent from Alices to  Bob.

This protocol is as follows:

* Alice performs a CNOT on the pair $(A,Q)$ being $Q$ the control and the entangled qubit $A$ the target. Then she performs a Hadamard gate on $Q$.

* Alice then measures both qubits with respect some basis, and transmits the classical outcomes to bob. Let's say that the classical bit emerging from measuring $A$ is $a$, and the classical bit from measuring $Q$ is $q$.

* Bob receives $a$ and $q$ and then he performs some operations depending on these values:

-if $a = 1$ he performs an X-gate (bit-flip) on qubit $B$

-if $b = 1$ he performs a Z-gate (phase-flip) on qubit $B$

This is, depending on the classical bit $aq$ being $00$, $01$, $10$ or $11$ he performs $I$, $Z$, $X$ or $ZX$. 

This procedure make qubit $B$ to be whatever qubit $Q$ was before the protocol.

Qubit $Q$ is not cloned, since after the protocol its state is now $|q \rangle$ since it has been measured.

The cost of this protocol is that the entanglement between $A$ and $Q$ has been eliminated.

$\textbf{1.2 Analysis}$

Let's analyze the state of the whole system in the different stages of the circuit:



Before operating any qubit the state of the system is $$|\pi_0 \rangle = |\phi^+ \rangle \otimes |Q \rangle$$

Consider the case where $|Q \rangle = \alpha|0 \rangle + \beta|1 \rangle$. Then: $$|\pi_0 \rangle = |\phi^+ \rangle \otimes (\alpha|0 \rangle + \beta|1 \rangle) = \frac{\alpha}{\sqrt{2}}|000 \rangle + \frac{\beta}{\sqrt{2}}|001 \rangle + \frac{\alpha}{\sqrt{2}}|110 \rangle + \frac{\beta}{\sqrt{2}}|111 \rangle $$

Then, when Alice applies the CNOT gate, it takes the $Q$ part (the most right qubit) as the control and the target is the $A$ part (the state in the middle):

$$|\pi_1 \rangle = \frac{\alpha}{\sqrt{2}}|000 \rangle + \frac{\beta}{\sqrt{2}}|011 \rangle + \frac{\alpha}{\sqrt{2}}|110 \rangle + \frac{\beta}{\sqrt{2}}|101 \rangle $$

Then, when Hadamard gate is applied it transforms the most right qubit as $|0 \rangle$ into $|+ \rangle$, and $|1 \rangle$ into $|- \rangle$:

$$|\pi_2 \rangle = \frac{\alpha}{\sqrt{2}}|00 \rangle|+ \rangle  + \frac{\beta}{\sqrt{2}}|01 \rangle|- \rangle  + \frac{\alpha}{\sqrt{2}}|11 \rangle|+ \rangle + \frac{\beta}{\sqrt{2}}|10 \rangle|- \rangle $$

which turns into

$$|\pi_2 \rangle = \frac{\alpha}{2}|000 \rangle + \frac{\alpha}{2}|001 \rangle + \frac{\beta}{2}|010 \rangle -  \frac{\beta}{2}|011 \rangle + \frac{\alpha}{2}|111 \rangle + \frac{\alpha}{2}|110 \rangle + \frac{\beta}{2}|100 \rangle - \frac{\beta}{2}|101 \rangle $$


Using multilinearity of the tensor product we arrange this state as: $|\pi_2 \rangle = \frac{1}{2}(\alpha |0 \rangle + \beta |1 \rangle)|00 \rangle + \frac{1}{2}(\alpha |0 \rangle - \beta |1 \rangle)|01 \rangle + \frac{1}{2}(\alpha |1 \rangle + \beta |0 \rangle)|10 \rangle + \frac{1}{2}(\alpha |1 \rangle - \beta |0 \rangle)|11 \rangle$ 

This looks like the qubit $B$ has "earned" information from qubit $Q$ which at the beggining had this coefficients $\alpha$ and $\beta$ that qubit $B$ now already has. This happens since we are manipulating states so that this set up is convenient for us when performing measurements.

Let's analyze the possible results and their probabilities.

We saw that the state of qubit $Q$ was $\alpha|0 \rangle + \beta|1 \rangle$ at the begining. We want qubit $B$ to be in this state.

* The measurement of $00 \rangle$, i.e. of $aq = 00$, has a probability of $$p_{00} = |\frac{1}{2}(\alpha|0 \rangle + \beta|1 \rangle)|^2 = \frac{|\alpha|^2 + |\beta|^2}{4} = \frac{1}{4}$$

  The state of qubit $B$ at then end is just that one of $Q$: $$\alpha|0 \rangle + \beta|1 \rangle$$

  Hence, Bob has nothing else to do, or he just can apply $I$.

  The final whole state is then $$|\psi \rangle = (\alpha|0 \rangle + \beta|1 \rangle)|00 \rangle$$

* The measaurement of $|01 \rangle$, i.e. of $aq = 01$, has a probability of

  $$p_{01} = |\frac{1}{2}(\alpha|0 \rangle - \beta|1 \rangle)|^2 = \frac{|\alpha|^2 + |-\beta|^2}{4} = \frac{1}{4}$$

  The state of qubit $B$ at then end is: $$\alpha|0 \rangle - \beta|1 \rangle$$

  Hence, in order to get $Q's$ initial state we have to change the second term by a minus sign. This is achieved by applying a Z-gate to $B$ since $Z|1 \rangle = - |1 \rangle$: $$Z(\alpha|0 \rangle - \beta|1 \rangle) = \alpha|0 \rangle + \beta|1 \rangle$$
Then we have achieved our task. The final whole state is then $$|\psi \rangle = (\alpha|0 \rangle + \beta|1 \rangle)|01 \rangle$$

* The measaurement of $|10 \rangle$, i.e. of $aq = 10$, has a probability of

  $$p_{01} = |\frac{1}{2}(\alpha|1 \rangle + \beta|0 \rangle)|^2 = \frac{|\alpha|^2 + |\beta|^2}{4} = \frac{1}{4}$$

  The state of qubit $B$ at then end is: $$\alpha|1 \rangle + \beta|0 \rangle$$

  Hence, in order to get $Q's$ initial state we have to flip the states corresponding to the amplitudes. This is achieved by applying the X-gate since it acts as a NOT gate: $X|0 \rangle = |1 \rangle$, and viceversa:
   $$X(\alpha|1 \rangle + \beta|0 \rangle) = \alpha|0 \rangle + \beta|1 \rangle$$
Then we have achieved our task. The final whole state is then $$|\psi \rangle = (\alpha|0 \rangle + \beta|1 \rangle)|10 \rangle$$

* The measaurement of $|11 \rangle$, i.e. of $aq = 11$, has a probability of

  $$p_{01} = |\frac{1}{2}(\alpha|1 \rangle - \beta|0 \rangle)|^2 = \frac{|\alpha|^2 + |-\beta|^2}{4} = \frac{1}{4}$$

  The state of qubit $B$ at then end is: $$\alpha|1 \rangle - \beta|0 \rangle$$

  Hence, in order to get $Q's$ initial state we have first to flip the states corresponding to the amplitudes, and then chance the sign of the term corresponding to the resultant $\beta|1 \rangle$. This is achieved by first applying the X-gate, and then by applying the Z-gate:
   $$ZX(\alpha|1 \rangle - \beta|0 \rangle) = Z(\alpha|0 \rangle - \beta|1 \rangle) = \alpha|0 \rangle + \beta|1 \rangle$$
Then we have achieved our task. The final whole state is then $$|\psi \rangle = (\alpha|0 \rangle + \beta|1 \rangle)|11 \rangle$$

$$\textbf{Qiskit Implementation}$$

For this implementation we will initialize the qubit $Q$ to transport to be in a superposition in the state $|+ \rangle$

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import array_to_latex
```

A quantum register is a collection of qubits on which gates can act. A classical register is a collection of bits that can be written and read within the coherence time of the quantum circuit.

```python
qubit = QuantumRegister(1,"Q")
ebit0 = QuantumRegister(1, "A") #(number of qubits in the register, name of the register)
ebit1 = QuantumRegister(1,"B")
q = ClassicalRegister(1,"q")
a = ClassicalRegister(1,"a")
```

```python
protocol = QuantumCircuit(qubit,ebit0,ebit1,q,a)
```

```python
# entangling the ebits for teleportation:
protocol.h(qubit) #qubit Q is initially in a superposition
protocol.h(ebit0)
protocol.cx(ebit0,ebit1)
protocol.barrier()
```

```python
#Alice's operations
protocol.cx(qubit,ebit0)
protocol.h(qubit)
protocol.barrier()
```

```python
#Alice measures and sends classical bits to Bob
protocol.measure(ebit0,a)
protocol.measure(qubit,q)
protocol.barrier()
```

```python
#Bob uses classical bits to apply gates:
with protocol.if_test((a, 1)): #if_test applies an operation conditionally depending on a classical bit or register
    protocol.x(ebit1)
with protocol.if_test((q, 1)):
    protocol.z(ebit1)
```

```python
protocol.draw('mpl')
```

Now let's make the simulation and get the histogram

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.result import marginal_counts
from qiskit import transpile

backend = AerSimulator()
protocol_trans = transpile(protocol,backend)

counts = AerSimulator().run(protocol_trans,shots=100000).result().get_counts()
plot_histogram(counts)
```

We can see the resulting states for the qubits A and Q, where qubit B is in the initial superposed Q state. 

As we saw, the state of qubit $B$ is a superposition, it has not been measured, in contrast with qubits $A$ and $Q$.

Now since Bob is in posession of superposed qubit $B$ he can perform a measurement on it and we can get statistics of the three qubits:

```python
# we add a classical register for bob measurement result
b = ClassicalRegister(1,"b")
protocol.add_register(b)
protocol.measure(ebit1,b)
protocol.draw('mpl')
```

```python
backend = AerSimulator()
protocol_trans = transpile(protocol,backend)

counts = AerSimulator().run(protocol_trans,shots=100000).result().get_counts()
plot_histogram(counts)
```

We see the lowest value has both 0 and 1, each corresponging when the superposed qubit $B$ is measured in either 0 or 1. The other two values correspond to the values for $Q$ and $A$.

Now we are able to see the measurements on the teleported state.

We can also implement this cirucit considering the initial qubit $Q$ to be entangled with some system $R$ such that their state is $$|QR \rangle = \alpha|0 \rangle_Q |\gamma_0 \rangle_R + \beta|1 \rangle_Q |\gamma_1 \rangle_R $$

Then we implement the circuit by the same way:



We'll work with the notation order BRAQ:

at the begining: $$|\pi_0 \rangle = |\phi^+ \rangle \otimes (\alpha|0 \rangle_Q |\gamma_0 \rangle_R + \beta|1 \rangle_Q |\gamma_1 \rangle_R)$$

```python
# create a new circuit including the same bits and qubits along with the auxiliary qubit R
aux = QuantumRegister(1,"R")
test = QuantumCircuit(aux,qubit,ebit0,ebit1,q,a,b)

# entangle Q with R
test.h(aux)
test.cx(aux,qubit)
test.barrier()

# recall the protocol circuit and combine with this test one. qubits in the argument tells that the protocol should
# operate on the qubits numbered 1,2, and 3, skipping qubit 0 which is R according to the arguments in test circuit.
test = test.compose(protocol,qubits=[1,2,3])
```

```python
test.draw('mpl')
```

```python
test_trans = transpile(test,backend)
```

```python
counts_test = backend.run(test_trans,shots=100000).result().get_counts()
```

```python
plot_histogram(counts_test)
```

As we can see, the lowest value is either 0 or 1 when measuring qubit $B$ since the initial entanglement between $Q$ and $R$ is teleported to $B$ and hence it is now entangled with $R$. This is purely due to the entanglement, not to a superposition.
